```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("matala2.csv")
```

```python
data
```

```python
import datetime
data['event_time']= pd.to_datetime(data['event_time'])
data.dtypes
```

```python
data=data.sort_values(by=['user_id','event_time','user_session'],ascending=True)
data=data.reset_index(drop=True)
data[['event_time','user_id','user_session']].head(50)
```

Q1

```python
data['duration_to_next_event'] = (data['event_time'].shift(-1)-data['event_time']).dt.seconds
data.loc[data['user_session'].shift(-1) !=data['user_session'], 'duration_to_next_event']=0
data[['duration_to_next_event','user_session','event_time']]
```

Q2

```python
data.sort_values(['user_id','event_time'])
data['funnel_number']=(data['event_time']-data['event_time'].shift(1)).dt.days
data['funnel_number'] = data['funnel_number'].apply(lambda x:1 if x > 5  else 0)
data.loc[data['user_id'].shift(1) !=data['user_id'], 'funnel_number']=1
data['funnel_number'] = data.groupby('user_id')['funnel_number'].cumsum()
data['funnel_number']
```

```python
data['funnel_number'].value_counts()
```

Q3

```python
data['index_in_funnel'] = data['funnel_number'].apply(lambda x:1 if x > 1  else 1)
data['index_in_funnel'] = data.groupby('user_id')['index_in_funnel'].cumsum()
data[['index_in_funnel','user_id']].head(50)
```

Q4

```python
data['price'] = data['price'].replace('[^0-9\.]+', '', regex=True)
data['price'] = data['price'].astype(float)
```

Q5

```python
import matplotlib as plt
counts = data['event_type'].value_counts()
counts.plot(kind='bar')
```

Q6

```python
df=data.copy()
list_of_viewed= df.loc[df['event_type'] == 'view'].groupby('user_session')['product_id'].apply(list)
list_of_added_to_cart = df.loc[df['event_type'] == 'cart'].groupby('user_session')['product_id'].apply(list)
list_of_purchased=df.loc[df['event_type'] == 'purchase'].groupby('user_session')['product_id'].apply(list)
```

```python
session_data = pd.concat([list_of_viewed, list_of_added_to_cart,list_of_purchased], axis=1)
session_data.columns=['list_of_viewed','list_of_added_to_cart','list_of_purchased']  
```

```python
funnel_number=data.groupby('user_session')['funnel_number'].max()
index_in_funnel=data.groupby('user_session')['index_in_funnel'].max()
duration_to_next_event=data.groupby('user_session')['duration_to_next_event'].sum()
num_of_events=data.groupby('user_session')['user_session'].count()
```

```python
user_id=data[['user_id','user_session']]
user_id = user_id.drop_duplicates('user_session').set_index('user_session')
```

```python
session_data = pd.concat([session_data,funnel_number,index_in_funnel,duration_to_next_event,num_of_events,user_id], axis=1)
session_data=session_data.reset_index(drop=True)
```
